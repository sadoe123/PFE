```python
import psycopg2
import pandas as pd
import numpy as np
import json
import re
import warnings
warnings.filterwarnings('ignore')

conn = psycopg2.connect(
    host='onepilot_postgres', port=5432,
    database='onepilot_dev', user='onepilot',
    password='onepilot_secret'
)

query = """
    SELECT se.id, se.name, se.entity_type,
           se.business_domain, se.business_concept,
           se.entity_class, se.row_count, se.description,
           COALESCE(json_agg(ef.name ORDER BY ef.position)
                    FILTER (WHERE ef.name IS NOT NULL), '[]') AS field_names
    FROM source_entities se
    LEFT JOIN entity_fields ef ON ef.entity_id = se.id
    WHERE se.source_id = '85a0ef4b-d9af-494f-b24f-ff710c21ba43'
      AND se.entity_type IN ('table', 'view', 'materialized_view')
    GROUP BY se.id, se.name, se.entity_type, se.business_domain,
             se.business_concept, se.entity_class, se.row_count, se.description
    ORDER BY se.name
"""
df = pd.read_sql(query, conn)
df['field_names'] = df['field_names'].apply(
    lambda x: x if isinstance(x, list) else json.loads(x)
)
print(f'✅ {len(df)} entités chargées')
print('Domaines:', df['business_domain'].value_counts().to_dict())
```

## 2. Synonymes et normalisation

```python
# Dictionnaire de synonymes FR/EN étendu
SYNONYMS = {
    'customer':   ['client', 'partner', 'debtor', 'bpcustomer', 'cli', 'cust'],
    'invoice':    ['facture', 'bill', 'ar', 'soinvoice', 'billing'],
    'order':      ['commande', 'sorder', 'oohead', 'vbak', 'salesorder', 'po', 'so'],
    'supplier':   ['fournisseur', 'vendor', 'creditor', 'bpvendor'],
    'employee':   ['employe', 'staff', 'personnel', 'rh', 'hr', 'worker'],
    'salary':     ['salaire', 'payroll', 'paie', 'wage', 'remuneration'],
    'account':    ['compte', 'gl', 'ledger', 'glacc', 'finance'],
    'payment':    ['paiement', 'reglement', 'payt', 'remittance'],
    'product':    ['produit', 'article', 'item', 'itmmaster'],
    'stock':      ['inventaire', 'inventory', 'warehouse', 'entrepot'],
    'budget':     ['prevision', 'forecast', 'bdg'],
    'bank':       ['banque', 'bancaire', 'bnk', 'tresorerie', 'cash'],
    'transaction':['mouvement', 'trn', 'transaction', 'bnkflst'],
    'balance':    ['solde', 'bali', 'blngst'],
    'purchase':   ['achat', 'procurement', 'fournisseur'],
    'sales':      ['vente', 'revenue', 'ca', 'chiffre'],
}
SYNONYM_INDEX = {w: k for k, words in SYNONYMS.items() for w in [k] + words}

def normalize_name(name):
    name = name.lower()
    name = re.sub(r'^[a-z]{1,3}_', '', name)
    name = re.sub(r'([a-z])([A-Z])', r'\1_\2', name).lower()
    tokens = re.split(r'[_\-\s\.]+', name)
    return [t for t in tokens if len(t) > 1]

def build_text_enriched(row):
    """Texte enrichi avec synonymes — version améliorée pour BERT."""
    parts = []
    
    # Nom original
    parts.append(row['name'].lower())
    
    # Tokens normalisés + synonymes
    tokens = normalize_name(row['name'])
    resolved = [SYNONYM_INDEX.get(t, t) for t in tokens]
    parts.extend(tokens + resolved)
    
    # Pour chaque token résolu, ajouter TOUS ses synonymes
    for r in set(resolved):
        if r in SYNONYMS:
            parts.extend(SYNONYMS[r])
    
    # Domaine métier
    if row['business_domain'] and row['business_domain'] != 'Autre':
        domain = row['business_domain'].lower()
        parts.append(domain)
        # Synonymes du domaine
        domain_synonyms = {
            'finance': ['invoice', 'facture', 'account', 'payment', 'bank', 'budget'],
            'ventes':  ['order', 'customer', 'sales', 'revenue'],
            'achats':  ['purchase', 'supplier', 'vendor', 'order'],
            'rh':      ['employee', 'salary', 'staff', 'personnel'],
            'logistique': ['stock', 'warehouse', 'shipment', 'delivery'],
        }
        parts.extend(domain_synonyms.get(domain, []))
    
    # Concept métier
    if row['business_concept']:
        concept = row['business_concept'].lower()
        parts.append(concept)
        if concept in SYNONYMS:
            parts.extend(SYNONYMS[concept])
    
    # Champs importants
    for fname in (row['field_names'] or [])[:15]:
        ftokens = normalize_name(fname)
        fresolved = [SYNONYM_INDEX.get(t, t) for t in ftokens]
        parts.extend(ftokens + fresolved)
    
    # Description
    if row['description']:
        parts.append(str(row['description']).lower())
    
    return ' '.join(dict.fromkeys(parts))  # dédupliquer

df['text'] = df.apply(build_text_enriched, axis=1)

print('✅ Textes enrichis construits')
print('\nExemple GS_GLACC:')
row = df[df['name'] == 'GS_GLACC']
if len(row):
    print(row['text'].values[0][:300])
```

## 3. Ground Truth corrigé — basé sur DOMAINES

```python
# ✅ FIX MAJEUR : ground truth basé sur domaines, pas sur noms exacts
# Pour chaque requête : domaines pertinents + concepts attendus

GROUND_TRUTH_DOMAINS = {
    'facture client': {
        'domains':  ['Finance'],
        'concepts': ['Account', 'Log'],
        'keywords': ['glacc', 'bnk', 'stmbnk', 'bdg', 'cshfl', 'blngst'],
    },
    'commande fournisseur': {
        'domains':  ['Achats', 'Ventes'],
        'concepts': ['Order', 'OrderLine'],
        'keywords': ['cpw', 'blsn', 'acgrp', 'brow', 'grp'],
    },
    'employee salary': {
        'domains':  ['RH'],
        'concepts': ['Employee'],
        'keywords': ['ltusrtl', 'empldlgt', 'aaltusrtl'],
    },
    'solde compte bancaire': {
        'domains':  ['Finance'],
        'concepts': ['Account'],
        'keywords': ['glacc', 'blngst', 'bnkflst', 'bali'],
    },
    'budget previsionnel finance': {
        'domains':  ['Finance'],
        'concepts': ['Account'],
        'keywords': ['bdg', 'budget', 'rev'],
    },
    'transaction paiement': {
        'domains':  ['Finance'],
        'concepts': ['Account', 'Log'],
        'keywords': ['payt', 'trn', 'bnkflst', 'cshfl'],
    },
    'stock produit inventaire': {
        'domains':  ['Logistique', 'Referentiel'],
        'concepts': ['Product', 'Category'],
        'keywords': ['stck', 'inv', 'wh', 'product'],
    },
}

def is_relevant(row, query_gt):
    """Vérifie si une entité est pertinente pour une requête."""
    name_lower = str(row['name']).lower()
    domain     = str(row['business_domain'] or '')
    concept    = str(row['business_concept'] or '')
    
    # Critère 1 : bon domaine
    if domain in query_gt['domains']:
        return True
    
    # Critère 2 : bon concept
    if concept in query_gt['concepts']:
        return True
    
    # Critère 3 : keyword dans le nom
    if any(kw in name_lower for kw in query_gt['keywords']):
        return True
    
    return False

# Calculer les entités pertinentes pour chaque requête
print('📊 Entités pertinentes par requête (ground truth corrigé):')
for query, gt in GROUND_TRUTH_DOMAINS.items():
    relevant_mask = df.apply(lambda r: is_relevant(r, gt), axis=1)
    n_relevant = relevant_mask.sum()
    print(f'  "{query}" → {n_relevant} entités pertinentes')
    gt['relevant_ids'] = set(df[relevant_mask]['name'].tolist())

print('\n✅ Ground truth corrigé prêt')
```

## 4. TF-IDF Embeddings

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_similarity
import time

t0 = time.time()
vectorizer = TfidfVectorizer(
    analyzer='word', ngram_range=(1, 2),
    max_features=5000, sublinear_tf=True, min_df=1,
)
tfidf_matrix = vectorizer.fit_transform(df['text'])
svd = TruncatedSVD(n_components=min(384, tfidf_matrix.shape[1]-1), random_state=42)
tfidf_embeddings = normalize(svd.fit_transform(tfidf_matrix))
t1 = time.time()

print(f'✅ TF-IDF : shape={tfidf_embeddings.shape}, temps={t1-t0:.2f}s')
print(f'   Variance expliquée : {svd.explained_variance_ratio_.sum():.3f}')
```

## 5. BERT Embeddings (texte enrichi)

```python
from sentence_transformers import SentenceTransformer

print('🔄 Chargement BERT all-MiniLM-L6-v2...')
t0 = time.time()
model = SentenceTransformer('all-MiniLM-L6-v2')
print(f'✅ Modèle chargé en {time.time()-t0:.1f}s')

print('🔄 Calcul embeddings BERT sur textes enrichis...')
t0 = time.time()
bert_embeddings = model.encode(
    df['text'].tolist(),
    batch_size=64,
    show_progress_bar=True,
    normalize_embeddings=True
)
print(f'✅ BERT : shape={bert_embeddings.shape}, temps={time.time()-t0:.2f}s')
```

## 6. Fonctions de recherche et métriques

```python
def search_tfidf(query, top_k=20):
    tokens = normalize_name(query)
    resolved = [SYNONYM_INDEX.get(t, t) for t in tokens]
    # Enrichir la requête avec synonymes
    extra = []
    for r in set(resolved):
        if r in SYNONYMS:
            extra.extend(SYNONYMS[r])
    query_text = ' '.join(resolved + tokens + extra + [query.lower()])
    
    q_vec = normalize(svd.transform(vectorizer.transform([query_text])))
    scores = cosine_similarity(q_vec, tfidf_embeddings)[0]
    top_idx = scores.argsort()[::-1][:top_k]
    return df.iloc[top_idx][['name','business_domain','business_concept']].copy().assign(
        score=scores[top_idx].round(4), method='TF-IDF'
    )

def search_bert(query, top_k=20):
    # ✅ FIX : enrichir la requête BERT aussi avec synonymes
    tokens = normalize_name(query)
    resolved = [SYNONYM_INDEX.get(t, t) for t in tokens]
    extra = []
    for r in set(resolved):
        if r in SYNONYMS:
            extra.extend(SYNONYMS[r])
    query_enriched = query + ' ' + ' '.join(set(resolved + extra))
    
    q_vec = model.encode([query_enriched], normalize_embeddings=True)
    scores = cosine_similarity(q_vec, bert_embeddings)[0]
    top_idx = scores.argsort()[::-1][:top_k]
    return df.iloc[top_idx][['name','business_domain','business_concept']].copy().assign(
        score=scores[top_idx].round(4), method='BERT'
    )

def precision_at_k(results, relevant, k):
    hits = sum(1 for n in results['name'].head(k) if n in relevant)
    return hits / k

def ndcg_at_k(results, relevant, k):
    top_k = results['name'].head(k).tolist()
    dcg  = sum(1/np.log2(i+2) for i,r in enumerate(top_k) if r in relevant)
    idcg = sum(1/np.log2(i+2) for i in range(min(len(relevant), k)))
    return dcg/idcg if idcg > 0 else 0

def mrr(results, relevant):
    for i, name in enumerate(results['name']):
        if name in relevant:
            return 1/(i+1)
    return 0

print('✅ Fonctions prêtes')
```

## 7. Comparaison qualitative — résultats côte à côte

```python
for query, gt in GROUND_TRUTH_DOMAINS.items():
    print(f'\n{"="*70}')
    print(f'🔍 "{query}"')
    print(f'   Domaines attendus : {gt["domains"]} | Concepts : {gt["concepts"]}')
    print(f'{"="*70}')
    
    rt = search_tfidf(query, top_k=5)
    rb = search_bert(query, top_k=5)
    
    print(f'{"Rk":<3} {"TF-IDF":<30} {"Dom":<12} {"Score":<7} | {"BERT":<30} {"Dom":<12} {"Score"}')
    print('-'*95)
    for i in range(5):
        t = rt.iloc[i]
        b = rb.iloc[i]
        t_ok = '✅' if t['name'] in gt['relevant_ids'] else '  '
        b_ok = '✅' if b['name'] in gt['relevant_ids'] else '  '
        print(f'{i+1:<3} {t_ok}{t["name"]:<28} {str(t["business_domain"]):<12} {t["score"]:<7.4f} | '
              f'{b_ok}{b["name"]:<28} {str(b["business_domain"]):<12} {b["score"]:.4f}')
```

## 8. Métriques corrigées

```python
metrics = []
for query, gt in GROUND_TRUTH_DOMAINS.items():
    relevant = gt['relevant_ids']
    rt = search_tfidf(query, top_k=20)
    rb = search_bert(query, top_k=20)
    
    metrics.append({
        'query':       query,
        'n_relevant':  len(relevant),
        'tfidf_p5':    precision_at_k(rt, relevant, 5),
        'bert_p5':     precision_at_k(rb, relevant, 5),
        'tfidf_p10':   precision_at_k(rt, relevant, 10),
        'bert_p10':    precision_at_k(rb, relevant, 10),
        'tfidf_ndcg':  ndcg_at_k(rt, relevant, 10),
        'bert_ndcg':   ndcg_at_k(rb, relevant, 10),
        'tfidf_mrr':   mrr(rt, relevant),
        'bert_mrr':    mrr(rb, relevant),
    })

mdf = pd.DataFrame(metrics)

print(f'{"Query":<35} {"Rel":<5} {"P@5 T|B":<14} {"NDCG T|B":<14} {"MRR T|B"}')
print('-'*80)
for _, r in mdf.iterrows():
    print(f'{r["query"][:33]:<35} {r["n_relevant"]:<5} '
          f'{r["tfidf_p5"]:.2f}|{r["bert_p5"]:.2f}     '
          f'{r["tfidf_ndcg"]:.2f}|{r["bert_ndcg"]:.2f}     '
          f'{r["tfidf_mrr"]:.2f}|{r["bert_mrr"]:.2f}')

print('-'*80)
print(f'{"MOYENNE":<40} '
      f'{mdf["tfidf_p5"].mean():.2f}|{mdf["bert_p5"].mean():.2f}     '
      f'{mdf["tfidf_ndcg"].mean():.2f}|{mdf["bert_ndcg"].mean():.2f}     '
      f'{mdf["tfidf_mrr"].mean():.2f}|{mdf["bert_mrr"].mean():.2f}')
```

## 9. Graphiques comparatifs

## 10. t-SNE — Visualisation des clusters

```python
from sklearn.manifold import TSNE
import matplotlib.cm as cm
from matplotlib.patches import Patch

print('🔄 Calcul t-SNE...')

domains = df['business_domain'].fillna('Autre').tolist()
unique_domains = sorted(set(domains))
cmap = cm.Set1(np.linspace(0, 1, len(unique_domains)))
dc = {d: cmap[i] for i, d in enumerate(unique_domains)}
pc = [dc[d] for d in domains]

fig, axes = plt.subplots(1, 2, figsize=(20, 8))
fig.suptitle("t-SNE — Clusters sémantiques TF-IDF vs BERT (coloré par domaine)",
             fontsize=13, fontweight='bold')

for ax, emb, title in [
    (axes[0], tfidf_embeddings, 'TF-IDF + SVD'),
    (axes[1], bert_embeddings,  'BERT all-MiniLM-L6-v2'),
]:
    tsne = TSNE(n_components=2, random_state=42, perplexity=30,
                max_iter=1000, init='pca', learning_rate='auto')
    coords = tsne.fit_transform(emb)
    
    ax.scatter(coords[:,0], coords[:,1], c=pc, alpha=0.6, s=15)
    ax.set_title(title, fontsize=11)
    ax.set_xlabel('t-SNE 1'); ax.set_ylabel('t-SNE 2')
    ax.legend(
        handles=[Patch(facecolor=dc[d], label=d) for d in unique_domains],
        loc='upper right', fontsize=8
    )

plt.tight_layout()
plt.savefig('/home/jovyan/work/tsne_corrected.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Sauvegardé : tsne_corrected.png')
```

## 11. Conclusion finale

```python
t_p5   = mdf['tfidf_p5'].mean()
b_p5   = mdf['bert_p5'].mean()
t_ndcg = mdf['tfidf_ndcg'].mean()
b_ndcg = mdf['bert_ndcg'].mean()
t_mrr  = mdf['tfidf_mrr'].mean()
b_mrr  = mdf['bert_mrr'].mean()

bert_wins = sum([b_p5 > t_p5, b_ndcg > t_ndcg, b_mrr > t_mrr])

print('='*65)
print('📊 RÉSUMÉ FINAL (ground truth corrigé)')
print('='*65)
print(f'\n{"Métrique":<15} {"TF-IDF":<10} {"BERT":<10} {"Gagnant":<10} {"Gain"}')
print('-'*55)

for name, tv, bv in [("Precision@5",t_p5,b_p5),("NDCG@10",t_ndcg,b_ndcg),("MRR",t_mrr,b_mrr)]:
    winner = '🤖 BERT' if bv > tv else '📊 TF-IDF'
    gain = f'+{(bv-tv)/max(tv,0.001)*100:.0f}%' if bv > tv else f'{(bv-tv)/max(tv,0.001)*100:.0f}%'
    print(f'{name:<15} {tv:<10.3f} {bv:<10.3f} {winner:<10} {gain}')

print('\n' + '='*65)
if bert_wins >= 2:
    improvement = (b_ndcg - t_ndcg) / max(t_ndcg, 0.001) * 100
    print(f'✅ RECOMMANDATION : Migrer vers BERT')
    print(f'   → BERT améliore le NDCG de +{improvement:.0f}%')
    print(f'   → Action : remplacer TF-IDF dans semantic_enricher.py')
    print(f'   → Coût : +80MB RAM, +2s au démarrage')
else:
    improvement = (t_ndcg - b_ndcg) / max(b_ndcg, 0.001) * 100
    print(f'✅ RECOMMANDATION : Garder TF-IDF')
    print(f'   → TF-IDF surpasse BERT de +{improvement:.0f}% sur ces données')
    print(f'   → Raison : noms cryptiques SXA (AA_, TH_, CS_) + synonymes bien configurés')
    print(f'   → Avantage : léger (pas de 80MB), rapide, déjà en prod')


print('   "Les noms de tables SXA étant cryptiques (AA_, TH_, CS_),"')
print('   "TF-IDF avec synonymes domaine-spécifiques surpasse BERT."')
print('   "Sur une base avec des noms lisibles (invoices, customers),"')
print('   "BERT serait supérieur."')
print('='*65)
```
